```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

import sympy as sp
sp.init_printing() 
from matplotlib import style
style.use('bmh')
plt.rcParams['figure.figsize'] = 16, 6
```

```python
#Define a varible in sympy
t = sp.symbols('t')
```

```python
s = 2*sp.exp(-t)*sp.cos(2*sp.pi*t)
s
```

- $\mathrm{d}s/\mathrm{d}t$

```python
s_der = sp.diff(s, t)
s_der
```

- $\int s(t) \mathrm{d}t$

```python
s_intg = sp.integrate(s, t)
s_intg
```

- Plot the three signals together.

```python
t = np.linspace(0, 2, 2000)
plt.figure(figsize=(16,6)) 
plt.plot(t, 10 * np.sin(2*np.exp(-t)*np.cos(2*np.pi*t)))
plt.plot(t, 10 * np.cos(2*np.exp(-t)*np.cos(2*np.pi*t)))
plt.plot(t, 10 * np.exp(2*np.exp(-t)*np.cos(2*np.pi*t)))
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.title(r'Plot signals')
plt.xlim([0, 2])
plt.grid(True)
plt.show()
```

- Fourier tranform of $S(w) = \mathcal{F} \left\{s(t) u(t)\right\},$ where $u(t)$ is the step function.

```python
w = sp.symbols('w')
t = sp.symbols('t')
u = 18*t**2 + 3*t + 5
sp.fourier_transform(s * u,t,w)
```

- Plot the real part of $S(w), \; -5 \leq w \leq 5$

```python
t = np.linspace(-5, 5, 2000)
plt.figure(figsize=(16,16)) 
plt.subplot(4,1,1); plt.plot(t, (2*np.exp(-t)*np.cos(2*np.pi*t)).real )
plt.ylabel('$\mathrm{Re}[x(t)]$')
plt.title(r'Real part of  $s(t)=2e^{t}\cos(2\pi t)$')
plt.grid(True)
plt.xlim([-6, 6])
plt.show()
```

- Plot the imaginary part of $S(w), \; -5 \leq w \leq 5$

```python
plt.figure(figsize=(16,16))
plt.subplot(4,1,2); plt.plot(t, (2*np.exp(-t)*np.cos(2*np.pi*t)).imag )
plt.ylabel('$\mathrm{Im}[x(t)]$')
plt.title(r'Imaginary part of  $s(t)=2e^{t}\cos(2\pi t)$')
plt.grid(True)
plt.xlim([-6, 6])
plt.show()
```

- Plot the absolute value of $S(w), \; -5 \leq w \leq 5$

```python
plt.figure(figsize=(16,16))
plt.subplot(4,1,3); plt.plot(t, np.abs(2*np.exp(-t)*np.cos(2*np.pi*t)) )
plt.ylabel(r'$|x(t)|$')
plt.title(r'Absolute value  of  $s(t)=2e^{t}\cos(2\pi t)$')
plt.grid(True)
plt.xlim([-6, 6])
plt.show()
```

- Plot the phase of $S(w), \; -5 \leq w \leq 5$

```python
plt.figure(figsize=(16,16))
plt.subplot(4,1,4); plt.plot(t, np.angle(2*np.exp(-t)*np.cos(2*np.pi*t)) )
plt.xlabel('$t$')
plt.ylabel(r'$\angle x(t)$')
plt.title(r'Phase value  of  $s(t)=2e^{t}\cos(2\pi t)$');
plt.grid(True)
plt.xlim([-6, 6])
plt.show()
```
